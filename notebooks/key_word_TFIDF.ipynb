```python
import pandas as pd
```

```python
# Define the file path
file_path = 'data/cleaned_data.csv'

# Load the cleaned data into a DataFrame
cleaned_df = pd.read_csv(file_path)
```

```python
# Separate the 'published_at' column into separate 'date' and 'time' columns
cleaned_df[['date', 'time']] = cleaned_df['published_at'].str.split(' ', expand=True)

# Print the DataFrame to verify the new columns
cleaned_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')

# Fit and transform the cleaned text data
tfidf_matrix = tfidf.fit_transform(cleaned_df['content'])

# Get the feature names (keywords)
feature_content = tfidf.get_feature_names_out()

# Print the feature names
print(feature_content)
```

```python

# Initialize the TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')

# Fit and transform the cleaned text data
tfidf_matrix = tfidf.fit_transform(cleaned_df['title'])

# Get the feature names (keywords)
feature_title = tfidf.get_feature_names_out()

# Print the feature names
print(feature_title)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize the TfidfVectorizer for title and body separately
tfidf_title = TfidfVectorizer(max_features=1000, stop_words='english')
tfidf_body = TfidfVectorizer(max_features=1000, stop_words='english')

# Fit and transform the cleaned text data for title and body
tfidf_title_matrix = tfidf_title.fit_transform(cleaned_df['title'])
tfidf_body_matrix = tfidf_body.fit_transform(cleaned_df['content'])

# Compute the cosine similarity between the two matrices
similarity_scores = cosine_similarity(tfidf_title_matrix, tfidf_body_matrix)

# Print the similarity scores
print(similarity_scores)
```

```python
# Step 1: Combine 'title', 'description', and 'body' into a single column 'text'
cleaned_df['text'] = cleaned_df['title'] + ' ' + cleaned_df['description'] + ' ' + cleaned_df['content']
```

```python
# Create a new DataFrame with 'text' and 'date' columns
df_topic = cleaned_df[['text', 'date']]

# Print the new DataFrame
df_topic
```

```python
# Check for null values in the 'text' column
text_null_count = cleaned_df['text'].isnull().sum()

# Check for null values in the 'date' column
date_null_count = cleaned_df['date'].isnull().sum()

# Print the counts of null values
print("Null values in 'text' column:", text_null_count)
print("Null values in 'date' column:", date_null_count)
```

# Topic Modeling

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Create a CountVectorizer to convert the text data into a bag-of-words representation
# Drop rows with missing values in the 'text' column


# Create a CountVectorizer to convert the text data into a bag-of-words representation
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
X = vectorizer.fit_transform(df_topic['text'])

# Train the LDA model
num_topics = 10  # Specify the number of topics
lda_model = LatentDirichletAllocation(n_components=num_topics, max_iter=5, learning_method='online', learning_offset=50., random_state=0)
lda_model.fit(X)
```

```python
# Get the topic assignments for each document
topic_assignments = lda_model.transform(X)

# Add the topic assignments to the DataFrame
df_topic['topic'] = topic_assignments.argmax(axis=1)

# Now you can proceed with the remaining code to analyze the topic trends
```

```python

# Print the topics
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda_model.components_):
    print(f"Topic {topic_idx}:")
    print(" ".join([feature_names[i] for i in topic.argsort()[:-10 - 1:-1]]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have already trained the LDA model and have the topic assignments available

# Extract date information
df_topic['date'] = pd.to_datetime(df_topic['date'])

# Count occurrences of each topic for each date
topic_counts = df_topic.groupby([df_topic['date'].dt.date, 'topic']).size().reset_index(name='count')

# Plot the scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=topic_counts, x='date', y='topic', hue='count', palette='viridis')
plt.title('Topic Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Topic')
plt.xticks(rotation=45)
plt.show()
```
