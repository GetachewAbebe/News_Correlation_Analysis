```python
import pandas as pd
import re

# Assuming you have the data in a DataFrame called cleaned_df and the country list in a variable called country_list

# Create a regular expression pattern to match the countries of interest
country_pattern = r'\b(?:' + '|'.join(country_list) + r')\b'

# Apply the regular expression pattern to the content column to identify articles about the countries of interest
cleaned_df['country_mentioned'] = cleaned_df['content'].str.findall(country_pattern, flags=re.IGNORECASE)

# Flatten the list of mentioned countries
flat_country_list = [country for sublist in cleaned_df['country_mentioned'] for country in sublist]

# Create a DataFrame to count the occurrences of each country
country_counts_df = pd.DataFrame(flat_country_list, columns=['Country']).value_counts().reset_index()
country_counts_df.columns = ['Country', 'Mentions']

# Display the top countries mentioned in the content
top_countries_mentioned = country_counts_df.head(10)  # Change the number to display more or fewer countries
print(top_countries_mentioned)
```

```python
import matplotlib.pyplot as plt

# Assuming you have already computed the top countries mentioned and stored them in top_countries_mentioned

# Plot the bar chart
plt.figure(figsize=(10, 6))
top_countries_mentioned.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries Mentioned in the Content')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Set the x-axis ticks and labels
plt.xticks(range(len(top_countries_mentioned)), top_countries_mentioned['Country'])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import re

# Assuming you have the data in a DataFrame called df and the country list in a DataFrame called country_df with a column named "Country"
# Replace df and country_df with the actual names of your DataFrames if they're different

# Define the regions and their corresponding countries
regions = {
    'Africa': ['Africa'], 
    'US': ['USA', 'United States'], 
    'China': ['China'], 
    'EU': ['European Union', 'Germany', 'France', 'Italy', 'Spain', 'Greece', 'Netherlands', 'Belgium', 'Austria', 'Portugal', 'Sweden', 'Hungary', 'Czech Republic', 'Denmark', 'Finland', 'Slovakia', 'Ireland', 'Croatia', 'Lithuania', 'Slovenia', 'Latvia', 'Estonia', 'Cyprus', 'Luxembourg', 'Malta', 'Bulgaria', 'Romania', 'Poland'], 
    'Russia': ['Russia'], 
    'Ukraine': ['Ukraine'], 
    'Middle East': ['Middle East', 'Saudi Arabia', 'Iran', 'Iraq', 'Yemen', 'Syria', 'Israel', 'Jordan', 'United Arab Emirates', 'Turkey', 'Lebanon', 'Oman', 'Kuwait', 'Qatar', 'Bahrain']
}

# Create a regular expression pattern to match the countries and regions of interest
country_pattern = r'\b(?:' + '|'.join(country for countries in regions.values() for country in countries) + r')\b'

# Apply the regular expression pattern to the content column to identify articles about the countries and regions of interest
cleaned_df['country_mentioned'] = cleaned_df['content'].str.contains(country_pattern, flags=re.IGNORECASE, na=False)

# Get the websites that reported news content about the countries and regions of interest
relevant_websites = cleaned_df.loc[cleaned_df['country_mentioned'], 'source_name'].unique()

# Display the list of relevant websites
print(relevant_websites)
```

```python
import pandas as pd

# Assuming you have the data in a DataFrame called cleaned_df with the specified columns

# Group the data by website domain and sentiment, then apply descriptive statistics
website_sentiment_stats = cleaned_df.groupby(['source_name', 'title_sentiment']).size().unstack(fill_value=0)

# Apply descriptive statistics to the grouped data for each website
descriptive_stats_by_media = website_sentiment_stats.groupby('source_name').agg({'Positive': ['mean', 'median', 'var'], 'Neutral': ['mean', 'median', 'var'], 'Negative': ['mean', 'median', 'var']})

# Display the descriptive statistics for each website
print(descriptive_stats_by_media)
```

```python
import matplotlib.pyplot as plt


# Create a box plot to compare the impact of using mean/average and median
descriptive_stats_by_media.boxplot(column=['Positive', 'Neutral', 'Negative'], showmeans=True)

# Set the title and labels
plt.title('Comparison of Mean/Average and Median for Sentiment Scores')
plt.ylabel('Sentiment Score')
plt.xlabel('Sentiment Category')

# Display the plot
plt.show()
```
